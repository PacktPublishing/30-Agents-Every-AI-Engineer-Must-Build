## Simulation 1: Self-Architecting Agent
**Chapter Reference:** pp. 493–494, *"Autonomous agent evolution and adaptation"*

> The self-architecting agent seeks a* that maximizes expected performance P(a)
> subject to alignment constraints a ∈ C. (Ch. 17, p. 494)

This simulation demonstrates the **architecture registry** pattern: a centralized
catalog of pre-validated reasoning modules that self-evolving agents can compose from.
The key insight is that the highest-performing pipeline may violate alignment constraints,
forcing the agent to select a constrained optimum.

````python
# Simulation 1: Self-Architecting Agent — Architecture Registry Search
# Ref: Ch.17, pp. 493–494, 'Autonomous agent evolution and adaptation'
# Author: Imran Ahmad

@fail_gracefully(fallback_return={"name": "Fallback-CoT", "performance": 0.50},
                 context="Architecture Registry Search")
def run_architecture_search():
    sim = ArchitectureRegistrySimulator()

    # Display full registry
    ColorLogger.section("Architecture Registry — Full Catalog")
    for i, pipeline in enumerate(sim.get_registry()):
        status = "COMPLIANT" if pipeline["alignment_compliant"] else "VIOLATION"
        ColorLogger.info(
            f"[{i+1}] {pipeline['name']} | P(a) = {pipeline['performance']:.2f} "
            f"| {status} | {pipeline['notes']}"
        )

    # Run constrained search
    print()
    ColorLogger.section("Meta-Optimization: Constrained Architecture Search")
    winner = sim.search_optimal()

    # LLM commentary
    print()
    llm_analysis = MockLLM.generate("Analyze the architecture search results")
    ColorLogger.simulation(f"LLM Analysis: {llm_analysis}")

    return winner

optimal_pipeline = run_architecture_search()
````

## Simulation 3: Ethical Circuit Breaker & Drift Detection
**Chapter Reference:** pp. 495–496, *"Agent governance and self-regulation"*

> The "ethical circuit breaker" pattern provides a graduated response to violations:
> logging an alert, increasing human oversight, restricting autonomy to pre-approved
> actions, and finally halting operation. (Ch. 17, pp. 495–496)

This simulation generates synthetic behavioral data that progressively drifts from
baseline. At each phase, the Kolmogorov-Smirnov statistic is computed and the
appropriate escalation level is triggered. All four levels fire in sequence.

### Ethical Circuit Breaker — Escalation Ladder

```
  Behavioral Drift Detected (KS test / JS divergence)
            │
            ▼
  ┌─────────────────────┐
  │  Level 1: LOG ALERT │  ← Minor drift (KS < 0.15)
  └────────┬────────────┘
           │  drift increases
           ▼
  ┌─────────────────────────────┐
  │  Level 2: INCREASE OVERSIGHT │  ← Moderate drift (KS 0.15–0.30)
  └────────┬────────────────────┘
           │  drift persists
           ▼
  ┌──────────────────────────────┐
  │  Level 3: RESTRICT AUTONOMY  │  ← Significant drift (KS 0.30–0.50)
  └────────┬─────────────────────┘
           │  violations continue
           ▼
  ┌──────────────────────────┐
  │  Level 4: HALT OPERATION │  ← Severe drift (KS > 0.50)
  └──────────────────────────┘
```


````python
# Simulation 3: Ethical Circuit Breaker — Drift Detection & Graduated Response
# Ref: Ch.17, pp. 495–496, 'Agent governance and self-regulation'
# Author: Imran Ahmad

@fail_gracefully(fallback_return=[], context="Ethical Circuit Breaker")
def run_circuit_breaker():
    sim = EthicalCircuitBreakerSimulator()
    results = sim.run_full_cascade()

    # Summary table
    print()
    ColorLogger.section("Escalation Summary")
    print(f"{'Phase':<30} {'KS Stat':>10} {'Level':>8} {'Response':<25}")
    print("-" * 75)
    for r in results:
        print(f"{r['phase']:<30} {r['ks_statistic']:>10.4f} {r['response_level']:>8} "
              f"{r['response_name']:<25}")

    # LLM commentary
    print()
    llm_analysis = MockLLM.generate("Analyze the ethics drift detection results")
    ColorLogger.simulation(f"LLM Analysis: {llm_analysis}")

    return results

circuit_breaker_results = run_circuit_breaker()
````

## Simulation 4: Memory Consolidation Engine
**Chapter Reference:** pp. 496–498, *"Brain-inspired cognitive architectures"*

> The complementary learning systems theory posits that biological memory involves
> two systems: a fast-learning hippocampal system that stores specific episodes and
> a slow-learning neocortical system that extracts statistical regularities.
> (Ch. 17, p. 497)

This simulation runs a full consolidation cycle: episodic replay → pattern extraction
→ semantic memory transfer → pruning. The key demonstration is **analogical transfer**:
the "retry-with-backoff" pattern generalizes from API calls to database connections.

### 📖 From the Book: *When Sleeping Brains Write Code*

> **Note (p. 497):** In 1994, neuroscientists Matthew Wilson and Bruce McNaughton recorded neurons in a
> rat's hippocampus while it navigated a maze, then recorded again while the rat slept. The sleeping brain
> replayed the same neural sequences, compressed into rapid bursts called *sharp-wave ripples*. The rat was
> rehearsing its day. Subsequent research confirmed this replay is not a mere echo but an active transfer
> mechanism: the hippocampus "teaches" the neocortex by replaying experiences until the slower system
> extracts durable patterns. Block the replay, and consolidation fails.
>
> The memory consolidation process in Simulation 4 below — where a scheduled batch job reviews episodes,
> extracts patterns, and prunes consolidated memories — is a computational analog of what Wilson and
> McNaughton's rats were doing in their sleep. The agents are, quite literally, *dreaming*.


### Memory Consolidation — Complementary Learning Systems

```
  ┌─────────────────────────────────────────────────────────────┐
  │                   EPISODIC MEMORY                          │
  │          (Fast-learning hippocampal system)                 │
  │                                                             │
  │  Episode 1  Episode 2  Episode 3  ...  Episode N           │
  │     │          │          │                │                │
  └─────┼──────────┼──────────┼────────────────┼────────────────┘
        │          │          │                │
        ▼          ▼          ▼                ▼
  ┌─────────────────────────────────────────────────────────────┐
  │              CONSOLIDATION PROCESS                         │
  │    (Offline replay → Pattern extraction → Transfer)        │
  │                                                             │
  │    1. Replay episodes   ──→  2. Extract patterns           │
  │    3. Update semantic   ──→  4. Prune consolidated         │
  └─────────────────────────────┬───────────────────────────────┘
                                │
                                ▼
  ┌─────────────────────────────────────────────────────────────┐
  │                  SEMANTIC MEMORY                           │
  │         (Slow-learning neocortical system)                 │
  │                                                             │
  │  Pattern A (retry-with-backoff)  →  Generalizes to new     │
  │  Pattern B (escalation logic)    →  domains via analogical │
  │  Pattern C (caching strategy)    →  transfer               │
  └─────────────────────────────────────────────────────────────┘
```


````python
# Simulation 4: Memory Consolidation — Episodic → Semantic Transfer
# Ref: Ch.17, pp. 496–498, 'Brain-inspired cognitive architectures'
# Author: Imran Ahmad

@fail_gracefully(fallback_return={}, context="Memory Consolidation")
def run_memory_consolidation():
    sim = MemoryConsolidationSimulator()

    # Pre-consolidation state
    ColorLogger.section("Pre-Consolidation: Episodic Memory Store")
    episodes = sim.get_episodes()
    for ep in episodes:
        ColorLogger.info(f"[{ep['id']}] {ep['event']} → {ep['outcome']}")

    # Run consolidation
    print()
    result = sim.consolidate()

    # Post-consolidation state
    print()
    ColorLogger.section("Post-Consolidation: Memory State")
    state = sim.get_memory_state()
    ColorLogger.success(f"Episodic memories remaining: {state['episodic_count']}")
    ColorLogger.success(f"Semantic patterns learned: {state['semantic_count']}")

    print()
    ColorLogger.section("Semantic Memory — Extracted Patterns")
    for name, pattern in state["semantic_patterns"].items():
        ColorLogger.success(
            f"Pattern: '{name}' | Sources: {pattern['source_count']} episodes | "
            f"Generalization: {pattern['generalization']}"
        )

    # LLM commentary
    print()
    llm_analysis = MockLLM.generate("Analyze the memory consolidation results")
    ColorLogger.simulation(f"LLM Analysis: {llm_analysis}")

    return result

consolidation_results = run_memory_consolidation()
````
